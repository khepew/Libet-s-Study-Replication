# Tasks Functions

```python
def rp_of_task(participant, task, electrode):
  filtered_rows = data_EEG[(data_EEG['participant'] == participant) & (data_EEG['series'] == task) & (data_EEG['electrode'] == electrode)]
  extracted_values = filtered_rows.loc[:, -2500: 800]
  extracted_values = extracted_values.fillna(0)
  mean_row = extracted_values.mean(axis=0)
  return mean_row
```
